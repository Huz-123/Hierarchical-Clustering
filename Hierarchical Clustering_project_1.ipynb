```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hierarchical clustering costomer data.csv')
```

```python
df
```

```python
df = df.iloc[:, 3:5].values
```

```python
df
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize = (10,7))
plt.title('Customer Dendrogram')
dend = shc.dendrogram(shc.linkage(df, method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters =5, linkage='ward')
cluster.fit_predict(df)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df[:,0], df[:,1], c=cluster.labels_, cmap = 'rainbow')
```

```python
new_data = np.array([[10, 20], [80, 90], [50, 60], [20, 80], [90, 20]])
new_cluster = AgglomerativeClustering(n_clusters = 5, linkage='ward')
predictions = new_cluster.fit_predict(new_data)
print(predictions)
```
